```python
#Find interquartile range and outlier for average price
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sales_data_specifycol=pd.read_csv("Desktop/sales_data_with_discounts.csv",usecols=["Avg Price", "Total Sales Value", "Discount Rate (%)",
       "Discount Amount", "Net Sales Value"])   
print("Average price list for sales")
avg=pd.read_csv("Desktop/sales_data_with_discounts.csv",usecols=["Avg Price"])     
print(avg)
print(avg.shape[0])
#Detail about Average Price
print("Detail about average price")
print(avg.describe())
#InterQuartileRange
avgq1=sales_data_specifycol["Avg Price"].quantile(.25)
avgq3=sales_data_specifycol["Avg Price"].quantile(.75)
avgIQR=avgq3-avgq1
print("Interaquartile={}  Q1={}  Q3={}".format(avgIQR,avgq1,avgq3))
avglowerbound=avgq1-(1.5*(avgIQR))
avgupperbound=avgq3+(1.5*(avgIQR))
print("Lower bound value:",avglowerbound)
print("Upper bound value:",avgupperbound)
#outlier list
outlier_value=avg[(avg["Avg Price"]<avglowerbound)|(avg["Avg Price"]>avgupperbound)]
print("No.of outliers items are",outlier_value.shape[0])
print("Outlier list")
print(outlier_value)
# Remove outlier value
not_outlier_value=avg[(avg["Avg Price"]>avglowerbound)&(avg["Avg Price"]<avgupperbound)]
print("No.of items not outliers are:",not_outlier_value.shape[0])
print(" Data without Outlier")
print(not_outlier_value)
sns.boxplot(not_outlier_value)
```
