```python
#Logistic Regression
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
#Load the data
train_data=pd.read_csv("Downloads/Logistic Regression (1)/Logistic Regression/Titanic_train.csv")
test_data=pd.read_csv("Downloads/Logistic Regression (1)/Logistic Regression/Titanic_test.csv")
data=pd.concat([train_data,test_data])
#Exploratory Dataset
#Find the missing values:
print("Total missing Values in the train dataset:",train_data.isna().sum().sum())
print("Total missing Values in the test dataset:",test_data.isna().sum().sum())
print("Total missing Values in the  dataset:",data.isna().sum().sum())
#Handle the missing values in train dataset:
for i in train_data.select_dtypes(include="number").columns:
    train_data=train_data.fillna(value=train_data[i].mean())
print("Total missing values in the train dataset after filling:",train_data.isna().sum().sum())
#Handle the missing values in test dataset:
for i in test_data.select_dtypes(include="number").columns:
    test_data=test_data.fillna(value=test_data[i].mean())
print("Total missing values in the train dataset after filling:",test_data.isna().sum().sum())
for i in data.select_dtypes(include="number").columns:
    data=data.fillna(value=data[i].mean())
print("Total missing values in the train dataset after filling:",data.isna().sum().sum())
#find duplicates in train dataset
print("No.of duplicate in train datasets:",end=" ")
print(train_data.duplicated().sum())
#find duplicates in test data
print("No.of duplicate in test datasets:",end=" ")
print(test_data.duplicated().sum())
print(train_data.info())
print("No.of duplicate in  datasets:",end=" ")
print(data.duplicated().sum())
print(data.info())
```

```python
#Treatment for outliers/boxplot capping for train data
def remove_outlier(col):
    q1,q3=np.percentile(col,[25,75])
    IQR=q3-q1
    lowerbound=q1-(1.5*(IQR))
    upperbound=q3+(1.5*(IQR))
    return lowerbound,upperbound
for i in data.select_dtypes(include="number").columns:
    lower,upper=remove_outlier(data[i])
    data[i]=np.where(data[i]<lower,lower,data[i])
    data[i]=np.where(data[i]>upper,upper,data[i])
    
```

```python
#box plot after capping  train dataset
for i in data.select_dtypes(include="number").columns:
    sns.boxplot(data[i])
    plt.show()
```

```python
#Treatment for outliers/boxplot capping for test data
def remove_outlier(col):
    q1,q3=np.percentile(col,[25,75])
    IQR=q3-q1
    lowerbound=q1-(1.5*(IQR))
    upperbound=q3+(1.5*(IQR))
    return lowerbound,upperbound
for i in test_data.select_dtypes(include="number").columns:
    lower,upper=remove_outlier(test_data[i])
    test_data[i]=np.where(test_data[i]<lower,lower,test_data[i])
    test_data[i]=np.where(test_data[i]>upper,upper,test_data[i])
```

```python
#Build and validate model

x=data.drop(columns="Sex")
y=data["Sex"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=25)

```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
ns_pro=[0 for _ in range(len(y_test))]
log_reg=LogisticRegression(multi_class="ovr",random_state=0).fit(x_train_scaled,y_train)
y_score =log_reg.fit(x_train_scaled, y_train).predict_proba(x_test_scaled)
y_pro=log_reg.predict_proba(x_test_scaled)
y_pro1=y_pro[:,1]
print("The accuracy between dataset:")
print("the score value of logistic regression train data is:")
print(log_reg.score(x_train_scaled,y_train))
print("the score value of logistic regression test data is:")
print(log_reg.score(x_test_scaled,y_test))
```

```python
y_pred=log_reg.predict(x_test_scaled)
accuracy=accuracy_score(y_pred,y_test)
print(accuracy)
```
