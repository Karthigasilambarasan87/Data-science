```python
#chisquare test
import itertools
import numpy as np
import scipy.stats as stat
from scipy.stats import chi2
import matplotlib.pyplot as plt
import pandas as pd
#create dataframe
data={'Satisfaction':['Very Satisfied','Satisfied','Neutral','Unsatisfied','Very Unsatisfied','Total'],
      'Smart Thermostat':[50,80,60,30,20,120],
      'Smart Light':[70,100,90,50,50,360],
      'Total':[120,180,150,80,70,600]}
customer=pd.DataFrame(data)
print(customer)
#checking two datasets
customer_table=pd.crosstab(customer["Smart Thermostat"],customer["Smart Light"])
print(customer_table)
#observed value
observed_value=customer_table.values
print("The Observed Values are:")
print(observed_value)
#Expected Value
value=stat.chi2_contingency(customer_table)
print("The Expected Values are:")
print(value)
chi_stats,pvalue,ddof,expected=value
print("Chi Square statistics:",chi_stats)
print("P value:",pvalue)
print("Degree of freedom:",ddof)
print("Expected Value:",expected)
alpha=0.05
#calculate critical value
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print("The critical value:",critical_value)
#check accept or reject null hypothesis using critical value
print("check accept or reject null hypothesis using critical value\n")
if chi_stats>=critical_value:
     print("The chisquare stastistics value:{} is greater than or equal to critical value:{}".format(chi_stats,critical_value))
     print("So,Reject null Hypothesis,There is a relationship between two categorical variables\n")
else:
    print("The chisquare stastistics value:{} is less than critical value:{}\n".format(chi_stats,critical_value))
    print("So, Accept null Hypothesis,There is a no relationship between two categorical variables\")
#check accept or reject null hypothesis using p value 
print("check accept or reject null hypothesis using p value")
if pvalue<=alpha:
    print("The P value:{} is less than or equal to alpha value:{}".format(pvalue,alpha))
    print("So,Reject null Hypothesis,There is a relationship between two categorical variables")
else:
    print("The chisquare P value:{} is greater than or equal to alpha value:{}".format(pvalue,alpha))
    print("Accept null Hypothesis,There is a no relationship between two categorical variables")
```
